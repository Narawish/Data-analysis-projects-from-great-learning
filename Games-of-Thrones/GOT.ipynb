```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
battle = pd.read_csv('battles.csv')
```

## renaming the columns

```python
battle.rename(columns={'attacker_1':'primary_attacker',
                       'defender_1':'primary_defender'},
                       inplace=True)
```

```python
battle.head()
```

## Looking for the frequency of the king of attacker and the king of defender

```python
battle['attacker_king'].value_counts()
```

```python
battle['defender_king'].value_counts()
```

```python
battle['location'].value_counts()
```

## Visualizing the data

```python
fig, ax = plt.subplots(3,1)
sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(data=battle, x='attacker_king', y='attacker_size', ax=ax[0])
sns.barplot(data=battle, x='defender_king', y='defender_size', ax=ax[1])
sns.countplot(x=battle['attacker_king'], hue=battle['battle_type'], ax=ax[2])
plt.show()
```
